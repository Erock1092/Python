```
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

#gets the input player per game stats
def getPlayer():
  return input("Enter player name:").lower().split(' ')

player = getPlayer()

url = 'https://www.basketball-reference.com/players/{}/{}01.html'.format(player[1][0], player[1] + player[0][0:2]) #parses entry and formats to url where input player data is found

html = urlopen(url)

soup = BeautifulSoup(html)
name = soup.find("h1", itemprop = "name").get_text().lower().split(' ')

if(not name == player):
  url = url = 'https://www.basketball-reference.com/players/{}/{}{}.html'.format(player[1][0], player[1] + player[0][0:2], "02") # player has the same 2 first initials and last name 


tables = pd.read_html(url)
per_game = tables[0].dropna()
per_game
```
